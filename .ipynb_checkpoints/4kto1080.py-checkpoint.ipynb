```python
# coding: utf-8
import os
import sys
import random
import math
import json
import codecs
import numpy

ROOT_DIR = os.getcwd()


#with codecs.open("test.json", 'r', 'utf8') as data_file:    
#    data = json.load(data_file)

f = open('/Users/kkshmzmbp/11-projects/1903-urbanorchestra/99-finaldata/json/C0003.json')
data = json.loads(f.read())
f.close()


for i in range(len(list(data))):
    for j in list(data[i]):
        if j == 'x':
            data[i][j] = data[i][j]/4
            if(data[i][j] < 480) :
                #print("a")
                data[i]["xcategory"] = "1"
            if(480 <= data[i][j] < 960):
                #print("b")
                data[i]["xcategory"] = "2"
            if(960 <= data[i][j] < 1440):
                #print("c")
                data[i]["xcategory"] = "3"
            if(1440 <= data[i][j] < 1920):
                #print("d")
                data[i]["xcategory"] = "4"
        if j == 'y':
            data[i][j] = data[i][j]/4
            if(data[i][j] < 540) :
                data[i]["ycategory"] = "a"
            else:
                data[i]["ycategory"] = "b"
    
    if (data[i]["xcategory"] == "1") and (data[i]["ycategory"] =="a"):
        data[i]["category"] = "1a"
        
    if (data[i]["xcategory"] == "1") and (data[i]["ycategory"] =="b"):
        data[i]["category"] = "1b"
        
    if (data[i]["xcategory"] == "2") and (data[i]["ycategory"] =="a"):
        data[i]["category"] = "2a"
    
    if (data[i]["xcategory"] == "2") and (data[i]["ycategory"] =="b"):
        data[i]["category"] = "2b"
        
    if (data[i]["xcategory"] == "3") and (data[i]["ycategory"] =="a"):
        data[i]["category"] = "3a"
    
    if (data[i]["xcategory"] == "3") and (data[i]["ycategory"] =="b"):
        data[i]["category"] = "3b"
        
    if (data[i]["xcategory"] == "4") and (data[i]["ycategory"] =="a"):
        data[i]["category"] = "4a"
    
    if (data[i]["xcategory"] == "4") and (data[i]["ycategory"] =="b"):
        data[i]["category"] = "4b"
        
        
        
    if (data[i]["xcategory"] or data[i]["ycategory"]):
        #print("a")
        data[i].pop('xcategory', None)
        data[i].pop('ycategory', None)

            
with open('/Users/kkshmzmbp/11-projects/1903-urbanorchestra/99-finaldata/json/edit-C0003.json', 'w') as outfile:
    #json.dumps(data,outfile)
     json.dump(data, outfile, indent=2, separators=(',', ': '), sort_keys=True)
        
```
